```python
import pandas as pd

df = pd.read_csv('/content/Hospital_Data.csv', on_bad_lines='skip')  # Skips junk rows

df = df[['Hospital_Name', 'Location', 'Department', 'Doctors_Count',
         'Patients_Count', 'Admission_Date', 'Discharge_Date', 'Medical_Expenses']].dropna()

df['Doctors_Count'] = df['Doctors_Count'].astype(int)
df['Patients_Count'] = df['Patients_Count'].astype(int)
df['Medical_Expenses'] = pd.to_numeric(df['Medical_Expenses'], errors='coerce')
df.to_csv('clean_hospital_data.csv', index=False)
print(df.head())

```
